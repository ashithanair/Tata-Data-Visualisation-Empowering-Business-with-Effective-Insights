```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
os.getcwd()
```

```python
df=pd.read_excel("test.xlsx")
df.head()
```

```python
df.info()
```

```python
df.describe(include="all")
```

```python
#finding out any null values
pd.isnull(np.nan)
```

```python
#count od null values in each columns
pd.isnull(df).sum()
```

```python
#drop null values
df=df.dropna()
df
```

```python
#finding duplicates
df.duplicated().sum()
```

```python
#drop duplicate values
df=df.drop_duplicates()
df
```

```python
df.describe(include="all")
```

```python
df["Country"].unique()
```

```python
df.Country.nunique()
```

```python
#check whether quantity is less than 1
less_quantity=df[df["Quantity"]<1]
less_quantity
```

```python
#drop the rows whose quantity is less than 1
df.drop(less_quantity.index,inplace=True)
df
```

```python
#finding whether any CustomerIDs are less than 1  
df[df["CustomerID"]<1].sum()
```

```python
df["Revenue"]=df["Quantity"]*df["UnitPrice"]
df
```

```python
#drop the rows which have revenue less than 1
df.drop(df[df["Revenue"]<1].index,inplace=True)
df
```

```python
#finding profit
df["Profit"]=df.Revenue-df.UnitPrice
df.Profit
```

```python
df["Profit"].sum()
```

```python
df["Revenue"].sum()
```

```python
t1=df[["Country","Revenue"]]
t1
```

```python
total = t1.groupby(t1.Country).sum().sort_values('Revenue', ascending=False).head(10)
total
```

```python
total.plot(kind="barh",figsize=(5,4))
```

```python
total.drop(total.index[0],inplace=True)
total
```

```python
total.plot(kind="barh",figsize=(5,4))
```

```python
t2=df[["Country","Profit"]]
t2
```

```python
totalprofit=t2.groupby(t2.Country).sum().sort_values("Profit",ascending=False).head(10)
totalprofit
```

```python
totalprofit.drop([totalprofit.index[0]],inplace=True)
```
